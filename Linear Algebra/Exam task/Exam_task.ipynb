## Задание 1
Исследуйте систему векторов на линейную зависимость или независимость без использования Python (вручную):

а1 = (1, 8, -1)
а2 =  (-2, 3, 3)
a3 = (4, -11, 9)  

Подробно распишите решение, а в итоге сделайте вывод про то, являются ли векторы зависимыми или независимыми

***Определение линейной зависимости***:

Векторы $a_1, a_2, a_3$ линейно зависимы, если существуют такие коэффициенты $c_1, c_2, c_3$, не все из которых равны нулю, что выполняется:

$
c_1 a_1 + c_2 a_2 + c_3 a_3 = 0.
$

Если таких коэффициентов нет, то векторы линейно независимы.

---
***1. Составим уравнения:***

Подставим векторы $a_1, a_2, a_3$:

$
c_1 (1, 8, -1) + c_2 (-2, 3, 3) + c_3 (4, -11, 9) = (0, 0, 0).
$

Получим систему линейных уравнений:

$
\begin{aligned}
& \quad c_1 - 2c_2 + 4c_3 = 0, \\
& \quad 8c_1 + 3c_2 - 11c_3 = 0, \\
& \quad -c_1 + 3c_2 + 9c_3 = 0.
\end{aligned}
$


***2. Запишем систему в матричном виде:***

$$
\begin{bmatrix}
1 & -2 & 4 \\
8 & 3 & -11 \\
-1 & 3 & 9
\end{bmatrix}
\cdot 
\begin{bmatrix}
c_1 \\
c_2 \\
c_3
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0
\end{bmatrix}
$$

Чтобы определить, существуют ли нетривиальные решения ($c_1, c_2, c_3$ не равные нулю), нужно найти ранг матрицы

***3. Находим ранг матрицы***
Найдем определитель:

$
\text{det}
\begin{bmatrix}
1 & -2 & 4 \\
8 & 3 & -11 \\
-1 & 3 & 9
\end{bmatrix}
$


$$
\text{det} =
1 \cdot
\begin{vmatrix}
3 & -11 \\
3 & 9
\end{vmatrix}
-
(-2) \cdot
\begin{vmatrix}
8 & -11 \\
-1 & 9
\end{vmatrix}
+
4 \cdot
\begin{vmatrix}
8 & 3 \\
-1 & 3
\end{vmatrix}
$$

$
\text{det} = 1 \cdot 60 - (-2) \cdot 61 + 4 \cdot 27 = 60 + 122 + 108 = 290
$


Так как определитель матрицы $ \neq 0 $, ранг матрицы равен 3 (максимальный). Следовательно, система уравнений имеет только тривиальное решение $c_1 = c_2 = c_3 = 0$


***Вывод:***

Векторы $a_1 = (1, 8, -1)$, $a_2 = (-2, 3, 3)$, $a_3 = (4, -11, 9)$ линейно независимы

## Задание 2

При каких значениях параметра a система является линейно зависимой?  
Найдите решение без использования Python (вручную)

a1 = (1; 0; 2; -3),  
a2 = (2; 2; -4; a),  
a3 = (3; 1; 2; -5)

Для того чтобы система векторов $\mathbf{a}_1 = (1, 0, 2, -3)$, $\mathbf{a}_2 = (2, 2, -4, a)$, $\mathbf{a}_3 = (3, 1, 2, -5)$ была линейно зависимой, должно существовать нетривиальное решение уравнения:

$
c_1 \mathbf{a}_1 + c_2 \mathbf{a}_2 + c_3 \mathbf{a}_3 = \mathbf{0}
$

где $\mathbf{0}$ — нулевой вектор, и хотя бы один из коэффициентов $c_1, c_2, c_3$ отличен от нуля

Это эквивалентно тому, что матрица, составленная из этих векторов, должна иметь ранг, меньший числа векторов. 
Составим матрицу:

$
A = \begin{bmatrix}
1 & 2 & 3 \\
0 & 2 & 1 \\
2 & -4 & 2 \\
-3 & a & -5
\end{bmatrix}
$

Чтобы найти условие линейной зависимости, рассмотрим определитель подматрицы из первых трех строк и столбцов:

$
\det \begin{bmatrix}
1 & 2 & 3 \\
0 & 2 & 1 \\
2 & -4 & 2
\end{bmatrix} = 1(2 \cdot 2 - 1 \cdot (-4)) - 2(0 \cdot 2 - 1 \cdot 2) + 3(0 \cdot (-4) - 2 \cdot 2)  $
$
= 1(4 + 4) - 2(0 - 2) + 3(0 - 4) = 8 + 4 - 12 = 0$


Поскольку определитель равен нулю, это значит, что векторы $\mathbf{a}_1, \mathbf{a}_2, \mathbf{a}_3$ уже линейно зависимы независимо от значения $a$

Следовательно, система линейно зависима для любого значения параметра $a$. Дополнительное выражение не требуется, так как уже три вектора линейно зависимы

## Задание 10
Разделите данные на предикторы и целевую переменную. Обучите модель с помощью классической линейной регрессии. Линейную регрессию необходимо реализовать без готовых библиотек.

Оцените качество модели с помощью хотя бы двух метрик качества.

Распишите решение с помощью кода на языке Python и напишите текстом интерпретацию полученного вывода.

```python
X = data.drop(columns=["cnt"])  # все переменные, кроме целевой переменной
y = data["cnt"]  # целевая переменная
```

```python
# столбец единичных значений для смещения (bias)
X = np.c_[np.ones(X.shape[0]), X]

X = np.array(X)
y = np.array(y)

# находим коэффициенты линейной регрессии
beta = np.linalg.inv(X.T @ X) @ X.T @ y
```

```python
# получаем прогнозы
y_pred = X @ beta
```

```python
# Среднеквадратичная ошибка
mse = np.mean((y - y_pred) ** 2)

# Коэффициент детерминации R^2
ss_total = np.sum((y - np.mean(y)) ** 2)
ss_residual = np.sum((y - y_pred) ** 2)
r2 = 1 - (ss_residual / ss_total)

print(f"MSE: {mse}")
print(f"R^2: {r2}")
```
