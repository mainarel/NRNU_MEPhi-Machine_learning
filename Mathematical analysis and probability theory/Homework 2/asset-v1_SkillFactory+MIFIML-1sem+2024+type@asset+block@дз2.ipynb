**Задача 1**. (11 баллов)

В известном магазине техники представлено очень много экземпляров различной бытовой техники двух видов: с возможностью подключения к системе "Умный дом" и без такой возможности. Причем, известно, что эти два вида техники представлены в пропорции 1 : 3 соответственно. Случайным образом выбрали 7 экземпляров техники для розыгрыша среди участниц конкурса "Мисс Рукоделье - 2024".  Какова вероятность, что среди них не более двух экземпляров с возможностью подключения к системе "Умный дом"?

Подробно распишите решение и обоснуйте выбор распределения для моделирования сюжета задачи. Ответ округлите до трёх знаков.


Нужно найти вероятность того, что среди 7 выбранных экземпляров бытовой техники не будет более двух с возможностью подключения к системе "Умный дом". Будем использовать биномиальное распределение, поскольку:

1. У нас есть два исхода для каждого экземпляра (с возможностью подключения к системе "Умный дом" или без).
2. Каждое событие выбора независимое.
3. Вероятность выбора экземпляра с возможностью подключения — это фиксированная вероятность.

***Определим параметры биномиального распределения***
Из условия задачи мы знаем, что техника с возможностью подключения к системе "Умный дом" встречается в пропорции 1 : 3, то есть вероятность того, что выбранный экземпляр будет с возможностью подключения к системе "Умный дом", равна:

$p = \frac{1}{1+3} = \frac{1}{4} = 0.25$

Вероятность того, что выбранный экземпляр будет без такой возможности (то есть, обычная техника) будет:

$q = 1 - p = 0.75$

Обозначим количество экземпляров с возможностью подключения к системе "Умный дом" среди 7 выбранных за $ X $. Тогда $X $ подчиняется биномиальному распределению с параметрами $ n = 7 $ и $p = 0.25 $:

$X \sim \text{Binomial}(7, 0.25)$


```python
# Параметры биномиального распределения
n = 7  # количество выбранных экземпляров
p = 0.25  # вероятность того, что экземпляр с подключением к системе "Умный дом"

# Рассчитываем вероятность того, что среди 7 выбранных экземпляров не более двух с подключением
prob = binom.cdf(2, n, p)
print("Вероятность", prob)
```

**Задача 2** (30 баллов)

Дан закон распределения количества публикаций в которой группе телеграм-каналов:

$$
\begin{array}{|c|c|c|c|c|c|c|}
\hline X \text { (число публикаций) } & 0 & 1 & 2 & 3 & 4 & 5 \\
\hline \mathrm{P} \text { (вероятность) } & ? & 0,3 & 0,3 & 0,2 & 0,1 & 0,1 \\
\hline
\end{array}
$$

Найдите пропущенную вероятность для X = 0, математическое ожидание и дисперсию. Решите задачу "вручную" и с помощью python.

***Найдем пропущенную вероятность для $ X = 0 $***

Из таблицы видно, что вероятности для значений $X = 1, 2, 3, 4, 5 $ составляют:

$
P(X=1) = 0.3, \quad P(X=2) = 0.3, \quad P(X=3) = 0.2, \quad P(X=4) = 0.1, \quad P(X=5) = 0.1.
$

Поскольку сумма всех вероятностей для дискретного распределения должна быть равна 1, мы можем вычислить вероятность для $ X = 0 $, обозначив её как $ P(X=0) $, из уравнения:

$
P(X=0) + P(X=1) + P(X=2) + P(X=3) + P(X=4) + P(X=5) = 1.
$

Подставим известные значения:

$
P(X=0) + 0.3 + 0.3 + 0.2 + 0.1 + 0.1 = 1.
$

Решим это уравнение для $P(X=0) $:

$
P(X=0) = 1 - (0.3 + 0.3 + 0.2 + 0.1 + 0.1) = 1 - 1.0 = 0.
$

Таким образом, пропущенная вероятность для $X = 0$ равна $0$

***Найдем математическое ожидание:***

Математическое ожидание $ M(X) $для дискретного распределения рассчитывается по формуле:

$
M(X) = \sum_{i} X_i \cdot P(X_i),
$

где $X_i$ — возможные значения случайной величины, а $P(X_i)$ — соответствующие им вероятности

Подставим значения:

$
M(X) = 0 \cdot P(X=0) + 1 \cdot P(X=1) + 2 \cdot P(X=2) + 3 \cdot P(X=3) + 4 \cdot P(X=4) + 5 \cdot P(X=5)
$

Так как $P(X=0) = 0 $, у нас остается:

$
M(X) = 1 \cdot 0.3 + 2 \cdot 0.3 + 3 \cdot 0.2 + 4 \cdot 0.1 + 5 \cdot 0.1
$

Вычислим:

$
M(X) = 0.3 + 0.6 + 0.6 + 0.4 + 0.5 = 2.4
$

Математическое ожидание $M(X) = 2.4$

***Находим дисперсию:***

Дисперсия $ D(X)$ рассчитывается по формуле:

$
D(X) = \sum_{i} (X_i - M(X))^2 \cdot P(X_i)
$

$
D(X) = (0 - 2.4)^2 \cdot 0 + (1 - 2.4)^2 \cdot 0.3 + (2 - 2.4)^2 \cdot 0.3 + (3 - 2.4)^2 \cdot 0.2 + (4 - 2.4)^2 \cdot 0.1 + (5 - 2.4)^2 \cdot 0.1
$

$
D(X) = 0 + 0.588 + 0.048 + 0.072 + 0.256 + 0.676 = 1.64
$

Дисперсия $ D(X) = 1.64 $


```python
# Значения X и их вероятности
X_values = np.array([0, 1, 2, 3, 4, 5])
P_values = np.array([0, 0.3, 0.3, 0.2, 0.1, 0.1])

# Математическое ожидание
M_X = np.sum(X_values * P_values)

# Дисперсия
D_X = np.sum((X_values - M_X) ** 2 * P_values)

print("Математическое ожидание: ", M_X, "\nДисперсия: ", D_X)
```

**Задача 3** (25 баллов)


Компания приобрела 10 000 новых ноутбуков. Вероятность поломки одного ноутбука в течение месяца равна 0,0002. Предполагается, что поломки отдельных ноутбуков происходят независимо друг от друга.

**Вопрос 1**. Найдите вероятность того, что за месяц сломается ровно 1 ноутбук.
Решение объясните. (5 баллов)


Задача предполагает, что поломки ноутбуков происходят независимо друг от друга, и вероятность поломки одного ноутбука в течение месяца равна 0,0002. Данная задача - задача с биноминальным распределением:

- Количество испытаний $ n = 10\,000 $ (количество ноутбуков),
- Вероятность успеха (поломки) $ p = 0.0002 $,
- Количество успешных исходов (поломок) $ k = 1$

```python
# Параметры задачи
n = 10000  # количество ноутбуков
p = 0.0002  # вероятность поломки одного ноутбука
k = 1  # ожидаемое количество успехов
# Вопрос 1: Вероятность того, что за месяц сломается ровно 1 ноутбук
P_1 = binom.pmf(k, n, p)
print("Вероятность ", P_1)
```

**Вопрос 2**. Компания планирует ввести программу расширенной гарантии, по которой она будет бесплатно ремонтировать или заменять ноутбуки, сломавшиеся в течение месяца. Если общее количество поломок за месяц превысит 5 ноутбуков, компания понесет значительные дополнительные расходы.

а) Найдите вероятность того, что за месяц сломается более 5 ноутбуков.Решение объясните (10 баллов)

б) Если стоимость ремонта или замены одного ноутбука составляет 500 долларов, вычислите математическое ожидание общих расходов компании на поломки за один месяц. (10 баллов)

Функция binom.sf(5, n, p) вычисляет $P(X > 5)$, то есть вероятность того, что поломается больше 5 ноутбуков, это обратная кумулятивная вероятность (survival function), которая дает $1 - P(X \leq 5)$

```python
# a)
P_2 = binom.sf(5, n, p)
print("Вероятность ", P_2)
```

```python
# b)
# стоимость ремонта или замены одного ноутбука
cost_per_broken_laptop = 500

# Математическое ожидание количества поломок
expected_broken_laptops = n * p

# Математическое ожидание общих расходов

expected_total_expenses = expected_broken_laptops * cost_per_broken_laptop
print("Математическое ожидание общих расходов ", expected_total_expenses)
```
